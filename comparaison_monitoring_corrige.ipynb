```python
mask_diff = (data_old != data_new) & ~(data_old.isna() & data_new.isna())

diff_cells = []
for col in data_old.columns:
    diff_rows = data_old[col] != data_new[col]
    diff_rows &= ~(data_old[col].isna() & data_new[col].isna())  # ignore les NaN alignés
    if diff_rows.any():
        for i in data_old[diff_rows].index:
            diff_cells.append({
                'Row': i,
                'Column': col,
                'Old Value': data_old.at[i, col],
                'New Value': data_new.at[i, col]
            })

diff_df = pd.DataFrame(diff_cells)
print(diff_df.head())
print(f"Nombre total de différences détaillées : {len(diff_df)}")
```
