```python
import pdfplumber
import pandas as pd
import re

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.options.display.max_columns=None
pd.options.mode.chained_assignment = None

# chemin vers le fichier PDF
pdf_path = "nauka_23.pdf"

# fonction pour extraire le texte d'un intervalle de pages spécifiques
def extract_text_from_pdf(pdf_path, start_page, end_page):
    text_by_page = []
    with pdfplumber.open(pdf_path) as pdf:
        # On commence à la page start_page (page 114) et on termine à la page end_page (page 126)
        for page_num in range(start_page - 1, end_page):  # les indices de pages commencent à 0
            page = pdf.pages[page_num]
            text_by_page.append(page.extract_text())
    return text_by_page
```

```python
# ИНСТИТУТ ЭКОТЕХНОЛОГИЙ И ИНЖИНИРИНГА

# Extraire le texte des pages 23 à 65
pages = extract_text_from_pdf(pdf_path, 23, 65)

# Fusionner le texte extrait en une seule chaîne
full_text = "\n".join([p for p in pages if p])

# Expression régulière pour rechercher les noms des départements et laboratoires
pattern = re.compile(r"КАФЕДРА [А-ЯЁ ]+|ЛАБОРАТОРИЯ [А-ЯЁ ]+")

# Trouver toutes les correspondances pour les départements et laboratoires
matches = list(pattern.finditer(full_text))

# Créer une liste des données
rows = []
for i in range(len(matches)):
    start = matches[i].start()
    end = matches[i + 1].start() if i + 1 < len(matches) else len(full_text)
    name = matches[i].group()
    content = full_text[start:end].strip()
    institute = "ИНСТИТУТ ЭКОТЕХНОЛОГИЙ И ИНЖИНИРИНГА"  # Ajout du nom de l'institut
    rows.append((name, content, institute))

# Créer un DataFrame avec les titres des colonnes en anglais
dataEcotech = pd.DataFrame(rows, columns=["Department", "Content", "Institute"])

# Afficher des informations sur le DataFrame
dataEcotech.info()
```
