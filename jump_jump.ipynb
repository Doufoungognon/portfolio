```python
def transform_jump(df):
    # оставляем только бакалавров 
    df['Учебная группа'] = df['Учебная группа'].str[0]
    df=df[(df['Учебная группа'] == 'Б')]

    # оставляем и переименовываем нужные столбцы
    df = df[['Курс', 'ФИО', 'Номер ЛД', 'Канцелярская дата','Учебная группа']]
    df.columns = ['course', 'fio', 'ld', 'date','level']

    # в столбце date оставляем только год 
    
    # убираем все символы кроме цифр, после этого заменяем 
    try:
        df['ld'] = df['ld'].astype(str).str.replace(r'\D', '', regex=True).astype(int)
    except ValueError:
        pass
   
    # обрезаем пробелы в начале и в конце строки, приводим к нижнему регистру
    df['course'] = df['course'].str.strip().str.lower()
    df['fio'] = df['fio'].str.strip().str.lower()
    df['level'] = df['level'].str.strip().str.lower()
    
    # приводим столбец date к типу datetime
    df['date'] = pd.to_datetime(df['date'], dayfirst=True).dt.year
    
    df=df[(df['course'] == 'второй') | (df['course'] == 'третий') | (df['course'] == 'четвертый')]

    return df

trans_jump20_21=transform_jump(jump20_21)
trans_jump21_22=transform_jump(jump21_22)
trans_jump22_23=transform_jump(jump22_23)
```

```python
def transform_first(df):
    df=df.iloc[:,0:3]
    df.columns=['fio','course','level']
    
    # Обрезаем пробелы в начале и в конце строки, приводим к нижнему регистру
    df['course'] = df['course'].str.strip().str.lower()
    df['fio'] = df['fio'].str.strip().str.lower()
    df['level'] = df['level'].str.strip().str.lower()

    df=df[(df['level'] == 'бакалавриат') | (df['level'] == 'академический бакалавриат')]
    
    df=df[(df['course'] == 'первый')]

    return df
```

```python
# бакалавры, поступившие в 2020 году
def all_in_4(first1,jump2,jump3,jump4):
    stud_list4=list(first1['fio']) # список фио первокурсников 
    pervi = len(stud_list4) # 1 курс 2020 год
    vtoroi = len(jump2.query('(course == "второй") and (date == 2021) and (fio in @stud_list4)')) # 2 курс 2021 год фио есть в stud_list4
    treti = len(jump3.query('(course == "третий") and (date == 2022) and (fio in @stud_list4)')) # 3 курс 2022 год фио есть в stud_list4
    chetverti = len(jump4.query('(course == "четвертый") and (date == 2023) and (fio in @stud_list4)')) # 4 курс 2023 год фио есть в stud_list4
    stud_for_course = [pervi,vtoroi,treti,chetverti]
    return stud_for_course
stud_from2020=all_in_4(trans_first20_21,trans_jump20_21,trans_jump21_22,trans_jump22_23)
stud_from2020
```
