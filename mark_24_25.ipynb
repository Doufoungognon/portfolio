# Creation des dataframes ["data_croise_24", "data_croise_25"]

```python
# Trouver les ld_number communs entre data_24 et data_25
common_ld = set(data_24['ld_number']) & set(data_25['ld_number'])

#Filtrer chaque DataFrame pour ne garder que ces ld_number
data_croise_24 = data_24[data_24['ld_number'].isin(common_ld)]
data_croise_25 = data_25[data_25['ld_number'].isin(common_ld)]

# Vérification : s'assurer qu'ils contiennent le même nombre de ld_number uniques
assert data_croise_24['ld_number'].nunique() == data_croise_25['ld_number'].nunique(), "Problème d'alignement des ld_number"

# Afficher les infos pour vérifier le résultat
print(f"Taille de data_croise_24 : {data_croise_24.shape}")
print(f"Taille de data_croise_25 : {data_croise_25.shape}")
print(f"Nombre de ld_number uniques : {data_croise_24['ld_number'].nunique()}")
```

# Caclul des moyennes

## marks_final_24

```python
# Création de data_simple_24 en sélectionnant les colonnes spécifiques
data_simple_24 = data_croise_24[['ld_number', 'date', 'subject', 'marks_final']]


# Filtrer les lignes où 'date' correspond à l'une des dates spécifiées
dates_filter = ['20_21_semestr1', '20_21_semestr2', '21_22_semestr1', '21_22_semestr2', 
                '22_23_semestr1', '22_23_semestr2']

data_simple_24 = data_simple_24[data_simple_24['date'].isin(dates_filter)]
# Remplacer les valeurs manquantes par zéro dans tout le DataFrame
data_simple_24 = data_simple_24.fillna(0)

```

```python
# Dictionnaire de conversion pour 'marks_final'
conversion_dict = {
    'Хорошо': 5,
    'зачтено': 3,
    'Отлично': 4,
    'Неявка': 0,
    'Удовлетворительно': 3,
    'Неудовлетворительно': 2,
    'не зачтено': 2,
    'Неявка по ув.причине': 1
}

# Appliquer ce dictionnaire de conversion à la colonne 'marks_final' de data_simple_24
data_simple_24['marks_final'] = data_simple_24['marks_final'].map(conversion_dict)

# Vérification
data_simple_24.info()
```

```python
# Calculer la moyenne de 'marks_final' pour chaque 'ld_number'
data_simple_24_moyenne = data_simple_24.groupby('ld_number')['marks_final'].mean().reset_index()

# Vérifier le résultat
data_simple_24_moyenne.head()
```

## marks_final_25

```python
# Création de data_simple_25 en sélectionnant les colonnes spécifiques
data_simple_25 = data_croise_25[['ld_number', 'date', 'subject', 'marks_final']]


# Filtrer les lignes où 'date' correspond à l'une des dates spécifiées
dates_filter = ['20_21_semestr1', '20_21_semestr2', '21_22_semestr1', '21_22_semestr2', 
                '22_23_semestr1', '22_23_semestr2']

data_simple_25 = data_simple_25[data_simple_25['date'].isin(dates_filter)]
# Remplacer les valeurs manquantes par zéro dans tout le DataFrame
data_simple_25 = data_simple_25.fillna(0)
```

```python
# Dictionnaire de conversion pour 'marks_final'
conversion_dict = {
    'Хорошо': 5,
    'зачтено': 3,
    'Отлично': 4,
    'Неявка': 0,
    'Удовлетворительно': 3,
    'Неудовлетворительно': 2,
    'не зачтено': 2,
    'Неявка по ув.причине': 1
}

# Appliquer ce dictionnaire de conversion à la colonne 'marks_final' de data_simple_24
data_simple_25['marks_final'] = data_simple_25['marks_final'].map(conversion_dict)

# Vérification
data_simple_25.info()
```

```python
# Calculer la moyenne de 'marks_final' pour chaque 'ld_number'
data_simple_25_moyenne = data_simple_25.groupby('ld_number')['marks_final'].mean().reset_index()

# Vérifier le résultat
data_simple_25_moyenne.head()
```

## Difference des moyennes

```python
# Créer une nouvelle colonne pour identifier les deux ensembles de données
data_simple_24_moyenne['dataset'] = 'Data 24'
data_simple_25_moyenne['dataset'] = 'Data 25'

# Fusionner les deux DataFrames pour comparaison
data_comparaison = pd.concat([data_simple_24_moyenne[['marks_final', 'dataset']],
                              data_simple_25_moyenne[['marks_final', 'dataset']]])

# Création du boxplot comparatif
plt.figure(figsize=(8, 6))
sns.boxplot(x='dataset', y='marks_final', data=data_comparaison, palette='Set2')

# Ajouter un titre et des labels
plt.title('Comparaison des moyennes de marks_final entre Data 24 et Data 25', fontsize=14)
plt.xlabel('Dataset', fontsize=12)
plt.ylabel('Moyenne de marks_final', fontsize=12)

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
mean_24 = data_simple_24_moyenne['marks_final'].mean()
mean_25 = data_simple_25_moyenne['marks_final'].mean()

difference_of_means = mean_24 - mean_25
print(f"Разница средних значений : {difference_of_means}")
```

```python
correlation = data_simple_24_moyenne['marks_final'].corr(data_simple_25_moyenne['marks_final'])
print(f"Коэффициент корреляции Пирсона : {correlation}")
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(data_simple_24_moyenne['marks_final'], data_simple_25_moyenne['marks_final'])
print(f"Erreur quadratique moyenne (MSE) : {mse}")
```
